### 3.3 Reprobados y Destacados [0.9 Puntos, 0.3 cada DataFrame]

Contentos con su desempe침o, la Universidad les solicita que obtengan los alumnos quienes han reprobado el a침o y los que han tenido desempe침o de destacado. Para esto se considerar치 que un alumno ha reprobado el a침o cuando el promedio de notas es inferior a 4.0 en la Escala Chilena, y se considerar치 destacado si posee notas superiores a 5.8 en todas las asignaturas (o sea notas superiores a 5.8 en `math score`, `reading score` y `writing score`).

Cree una funci칩n que permita obtener tres Dataframes:
- El primero deber치 poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.
- El tercero ser치 un Data Frame con los alumnos destacados ordenados de mayor a menor.

**To-Do:**

- [ ] Obtener DataFrame con los alumnos que reprobaron el a침o.
- [ ] Obtener DataFrame similar al original, pero marcados con una "R" el atributo GPA de los alumnos reprobados.
- [ ] Obtener DataFrame con los alumnos destacados ordenados de mayor a menor.

**Respuesta:**

````python
def reprobados_destacados(dataframe_in):
    df_copy = dataframe_in.copy()
    mascara = df_copy.loc[:,'GPA'] < 4.00
    reprobados = df_copy.loc[mascara,:]
    df_copy.loc[:,'GPA'] = np.where(df_copy.loc[:,'GPA'] < 4.00, 'R', df_copy.loc[:,'GPA'])
    aprobados = df_copy.loc[~mascara,:].sort_values('GPA', ascending=False)
    
    return reprobados, df_copy, aprobados
````

Para el caso de evaluaci칩n sobre cu치l es el mejor join a utilizar para generar el nuevo dataframe, seg칰n los resultados arrojados por cada tipo de join, se opta por el left join. Esto porque permite conservar de mejor forma la integralidad de los datos. Es decir, en el proceso de uni칩n de las tablas, no perdemos ning칰n caso del dataframe df_grades  original (que se trabaj칩, limpi칩 y transform칩 con tanto trabajo y sentido anal칤tico), y se gana todos los datos de las nuevas columnas al momento de usar como llave los nombres de los estudiantes (aunque si en el segundo dataframe no hay concidencias en base a la llave utilizada, se generar치n datos NAN). Para el caso particular de este trabajo, el inner join arroja los mismos resultados que el left join, pues en ambas tablas hay coincidencias en base a la llave utilizada, raz칩n por la cual se conserva la integralidad de la tabla df_grades. Por las razones opuestas, se descarta el right y el outer join: se generan casos NAN en los atributos, perdiendose integralidad de los datos.
