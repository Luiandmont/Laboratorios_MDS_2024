````python
def clean_data(dataframe):
    dataframe.columns = dataframe.columns.str.replace(' ', '_')
    nums = [pd.to_numeric(dataframe[i]) for i in dataframe.columns[-3:]]
    nums = pd.DataFrame(nums).T
    categs = dataframe.iloc[:,:6].astype('category')
    df = pd.concat([categs,nums],axis=1)
    df = df.drop_duplicates(subset=['names'], keep='first')
    df = df.loc[(df.loc[:, "math_score"] <= 100) & (df.loc[:, "reading_score"] <= 100) & (df.loc[:, "writing_score"] <= 100)]
    
    return df.reset_index(drop=True)
````

Contentos con su desempeño, la Universidad les solicita que obtengan los alumnos quienes han reprobado el año y los que han tenido desempeño de destacado. Cree una función que permita obtener tres Dataframes:
- El primero deberá poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.
- El tercero será un Data Frame con los alumnos destacados ordenados de mayor a menor.

**To-Do:**

- [ ] Obtener DataFrame con los alumnos que reprobaron el año.
- [ ] Obtener DataFrame similar al original, pero marcados con una "R" el atributo GPA de los alumnos reprobados.
- [ ] Obtener DataFrame con los alumnos destacados ordenados de mayor a menor.

**Respuesta:**

````python
def reprobados_destacados(dataframe_in):
    df_copy = dataframe_in.copy()
    mascara = df_copy.loc[:,'GPA'] < 4.00
    reprobados = df_copy.loc[mascara,:]
    df_copy.loc[:,'GPA'] = np.where(df_copy.loc[:,'GPA'] < 4.00, 'R', df_copy.loc[:,'GPA'])
    aprobados = df_copy.loc[~mascara,:].sort_values('GPA', ascending=False)
    
    return reprobados, df_copy, aprobados
````

````python
# SE RELLENAN ESPACIOS EN LOS NOMBRES DE LAS COLUMNAS PARA EVITAR ERRORES O CAÍDAS DEL CÓDIGO.

dff.columns = dff.columns.str.replace(' ', '_')
dff.columns
````

````python
# CALCULO PROMEDIO DE NOTAS GPA

dff['GPA'] = dff[['math_score','reading_score','writing_score']].mean(axis=1).round(2)
````

````python
# SIMPLIFICACIÓN NIVEL EDUCACIONAL PADRES MEDIANTE MAPEO Y FUNCIÓN DE CONVERSIÓN.

def simplify_education_level(level):
    conversion_dict = {
        'some high school': 'school',
        'some college': 'school',
        'high school': 'school',
        "bachelor's degree": 'college',
        "associate's degree": 'college',
        "master's degree": 'postgraduate'
    }
    return conversion_dict.get(level, level)


dff['simple_parental_level_of_education'] = list(map(simplify_education_level, dff['parental_level_of_education']))
````

````python
t1 = dff.groupby(['race/ethnicity','simple_parental_level_of_education']).agg({'math_score':'mean','reading_score':'mean','writing_score':'mean','GPA':'mean'}).round(2).reset_index()
````

**RESPUESTA**  

Para el caso de evaluación sobre cuál es el mejor join a utilizar para generar el nuevo dataframe, según los resultados arrojados por cada tipo de join, se opta por el left join. Esto porque permite conservar de mejor forma la integralidad de los datos. Es decir, en el proceso de unión de las tablas, no perdemos ningún caso del dataframe df_grades  original (que se trabajó, limpió y transformó con tanto trabajo y sentido analítico), y se gana todos los datos de las nuevas columnas al momento de usar como llave los nombres de los estudiantes (aunque si en el segundo dataframe no hay concidencias en base a la llave utilizada, se generarán datos NAN). Para el caso particular de este trabajo, el inner join arroja los mismos resultados que el left join, pues en ambas tablas hay coincidencias en base a la llave utilizada, razón por la cual se conserva la integralidad de la tabla df_grades. Por las razones opuestas, se descarta el right y el outer join: se generan casos NAN en los atributos, perdiendose integralidad de los datos.

````python
# Inserte su respuesta aquí

dfo = pd.read_csv('other_grades.csv')
dfo
````

````python
# OUTER JOIN

df_out = pd.merge(
    left=df_grades,
    right=dfo,
    on="names",
    how="outer",
)

df_out
````

````python
# LEFT JOIN

df_left = pd.merge(
    left=df_grades,
    right=dfo,
    on="names",
    how="left",
)

df_left
````

````python
# RIGHT JOIN

df_rg = pd.merge(
    left=df_grades,
    right=dfo,
    on="names",
    how="right",
)

df_rg
````

````python
# INNER JOIN

df_in = pd.merge(
    left=df_grades,
    right=dfo,
    on="names",
    how="inner",
)

df_in
````
