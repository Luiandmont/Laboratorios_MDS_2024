**Respuestas**


**RESPUESTA 1**
   Apache Spark es un motor de procesamiento de datos en memoria distribuido de c√≥digo abierto que proporciona una API unificada para el procesamiento de grandes vol√∫menes de datos de manera distribuida. Algunas de sus principales ventajas sobre Pandas son:
   - Escalabilidad: Spark puede manejar conjuntos de datos mucho m√°s grandes que Pandas al distribuir el procesamiento en un cl√∫ster de m√°quinas.
   - Procesamiento en memoria: Spark realiza operaciones en memoria cuando es posible, lo que puede ofrecer un rendimiento superior en comparaci√≥n con Pandas para conjuntos de datos grandes.

**RESPUESTA 2**
   RDD (Resilient Distributed Dataset) es la abstracci√≥n fundamental de datos en Spark, que representa una colecci√≥n inmutable y distribuida de objetos. Una de sus principales caracter√≠sticas es la tolerancia a fallos, lo que significa que los RDDs pueden reconstruirse autom√°ticamente en caso de fallo de una partici√≥n o de un nodo.
   Los DataFrames de Spark, por otro lado, son una API de nivel superior que se construye sobre los RDDs y proporcionan una interfaz m√°s estructurada y optimizada para el procesamiento de datos. Los DataFrames de Spark son m√°s f√°ciles de usar y suelen ofrecer un mejor rendimiento que trabajar directamente con RDDs.

**RESPUESTA 3**
   - Transformaciones: Son operaciones que se aplican a un RDD o DataFrame para producir otro RDD o DataFrame nuevo. Las transformaciones son generalmente ***lazy***, lo que significa que no se ejecutan de inmediato, sino que se acumulan en un grafo de transformaciones hasta que se activa una acci√≥n.
     Ejemplo de transformaci√≥n: `filter`, `groupBy`.
   - Acciones: Son operaciones que se ejecutan en un RDD o DataFrame y desencadenan la computaci√≥n real para producir un resultado. Las acciones provocan que las transformaciones anteriores se ejecuten realmente.
     Ejemplo de acci√≥n: `show`, `count`.
   Cuando se ejecuta una acci√≥n, Spark genera un plan de ejecuci√≥n (llamado DAG) a partir del grafo de transformaciones acumuladas y lo env√≠a a los ejecutores para su ejecuci√≥n.

**RESPUESTA 4**
   El particionamiento en Spark es crucial para el rendimiento del procesamiento de datos, ya que afecta la distribuci√≥n y paralelizaci√≥n de las tareas en un cl√∫ster. El particionamiento determina c√≥mo se distribuyen los datos entre los nodos del cl√∫ster y puede afectar la eficiencia del procesamiento paralelo y la utilizaci√≥n de los recursos del cl√∫ster.
   Al particionar los datos adecuadamente, se puede lograr un equilibrio en la carga de trabajo entre los nodos del cl√∫ster, lo que mejora la utilizaci√≥n de los recursos y reduce los cuellos de botella.

**RESPUESTA 5**
   - Spark Driver: Es el proceso principal de una aplicaci√≥n Spark, responsable de mantener informaci√≥n sobre la topolog√≠a del cl√∫ster, distribuir el c√≥digo y los datos a los ejecutores, y coordinar la ejecuci√≥n de tareas en los ejecutores. El controlador es donde se inicia la aplicaci√≥n Spark y donde se ejecuta el c√≥digo del usuario.
   - Spark Executor: Son los procesos que realizan la ejecuci√≥n real de las tareas de una aplicaci√≥n Spark. Cada ejecutor se ejecuta en un nodo del cl√∫ster y es responsable de ejecutar un conjunto de tareas asignadas por el controlador. Los ejecutores son responsables de mantener los datos en memoria y realizar el procesamiento real de los datos de la aplicaci√≥n.

**RESPUESTA ?**
   El Catalyst Optimizer se encarga de optimizar y ejecutar consultas SQL en Spark SQL. Su funci√≥n es realizar optimizaciones en tiempo de ejecuci√≥n para mejorar el rendimiento de las consultas, excluyendo columnas no utilizadas, la ordenaci√≥n de operaciones y la generaci√≥n de planes de ejecuci√≥n m√°s eficientes.

Todas estas variables corresponden a un valor entero discreto; sin embargo, todas utilizan valores num√©ricos del tipo double (float) con decimales, lo que en estricto rigor, si bien no entorpecen

````python
# RESPUESTA N¬∞ 5

from pyspark.sql.types import StringType, IntegerType, FloatType

def cast_cols(df, cols_types):
    for col_name, new_type in cols_types.items():
        df = df.withColumn(col_name, df[col_name].cast(new_type))
    return df

cols_to_cast = {"Quantity": IntegerType(), "Customer Age": IntegerType(),"Is Fraudulent": IntegerType(),"Account Age Days": IntegerType(),"Transaction Hour":IntegerType()}
df = cast_cols(df, cols_to_cast)
df.printSchema()
````

````python
# RESPUESTA N¬∞ 1
import pyspark.sql.functions as F

df = df.withColumn('Transaction bp', F.log10(48 + df['Transaction Amount']**36))
df.show(5)
````

````python
# RESPUESTA N¬∞ 2
df = df.withColumn('Transaction Month',F.month(df['Transaction Date']))

df.show(5)
````

````python
# RESPUESTA N¬∞ 3

df = df.withColumn('Type_of_purchase', when(df['Quantity'] < 5, 'minorista').otherwise('mayorista'))

df = df.withColumn('Is Alien', when(df['Is Alien'] == 1, 'Alien').otherwise('Human'))

df.show(5)
````

````python
# RESPUESTA N¬∞ 4

df.select(['Is Alien','Type_of_purchase','Transaction ID']).filter((df['Is Alien'] == 'Alien') &
          (df['Type_of_purchase'] == 'mayorista')).show()
````

````python
# RESPUESTA N¬∞ 5

df.groupBy("Is Alien").agg({"Quantity":"sum"}).show()
````

````python
# RESPUESTA N¬∞ 6

pivot_df = df.groupBy("Payment Method").pivot("Is Alien").agg({"Transaction Amount": "sum"})
pivot_df.show()
````

Las funciones de ventana permiten realizar c√°lculos sobre un conjunto de filas que est√°n relacionadas con la fila actual. Por ejemplo, UNBOUNDED PRECEDING se usa para indicar que el rango de la funci√≥n de ventana comienza desde la primera fila de la partici√≥n o del conjunto de resultados, lo cual es √∫til para calcular sumas acumulativas hasta la fila actual. Las variaciones comunes de este uso incluyen:

- `UNBOUNDED PRECEDING` to `CURRENT ROW`: Calcula desde el inicio de la partici√≥n hasta la fila actual.
- `UNBOUNDED PRECEDING` to `UNBOUNDED FOLLOWING`: Cubre todas las filas dentro de la partici√≥n.
- `VALUE PRECEDING` to `VALUE FOLLOWING`: Establece un rango espec√≠fico basado en valores antes y despu√©s de la fila actual.
