**RESPUESTA 1**
   Apache Spark es un motor de procesamiento de datos en memoria distribuido de código abierto que proporciona una API unificada para el procesamiento de grandes volúmenes de datos de manera distribuida. Algunas de sus principales ventajas sobre Pandas son:
   - Escalabilidad: Spark puede manejar conjuntos de datos mucho más grandes que Pandas al distribuir el procesamiento en un clúster de máquinas.
   - Procesamiento en memoria: Spark realiza operaciones en memoria cuando es posible, lo que puede ofrecer un rendimiento superior en comparación con Pandas para conjuntos de datos grandes.

**RESPUESTA 2**
   RDD (Resilient Distributed Dataset) es la abstracción fundamental de datos en Spark, que representa una colección inmutable y distribuida de objetos. Una de sus principales características es la tolerancia a fallos, lo que significa que los RDDs pueden reconstruirse automáticamente en caso de fallo de una partición o de un nodo.
   Los DataFrames de Spark, por otro lado, son una API de nivel superior que se construye sobre los RDDs y proporcionan una interfaz más estructurada y optimizada para el procesamiento de datos. Los DataFrames de Spark son más fáciles de usar y suelen ofrecer un mejor rendimiento que trabajar directamente con RDDs.

**RESPUESTA 3**
   - Transformaciones: Son operaciones que se aplican a un RDD o DataFrame para producir otro RDD o DataFrame nuevo. Las transformaciones son generalmente ***lazy***, lo que significa que no se ejecutan de inmediato, sino que se acumulan en un grafo de transformaciones hasta que se activa una acción.
     Ejemplo de transformación: `filter`, `groupBy`.
   - Acciones: Son operaciones que se ejecutan en un RDD o DataFrame y desencadenan la computación real para producir un resultado. Las acciones provocan que las transformaciones anteriores se ejecuten realmente.
     Ejemplo de acción: `show`, `count`.
   Cuando se ejecuta una acción, Spark genera un plan de ejecución (llamado DAG) a partir del grafo de transformaciones acumuladas y lo envía a los ejecutores para su ejecución.

**RESPUESTA 4**
   El particionamiento en Spark es crucial para el rendimiento del procesamiento de datos, ya que afecta la distribución y paralelización de las tareas en un clúster. El particionamiento determina cómo se distribuyen los datos entre los nodos del clúster y puede afectar la eficiencia del procesamiento paralelo y la utilización de los recursos del clúster.
   Al particionar los datos adecuadamente, se puede lograr un equilibrio en la carga de trabajo entre los nodos del clúster, lo que mejora la utilización de los recursos y reduce los cuellos de botella.

**RESPUESTA 5**
   - Spark Driver: Es el proceso principal de una aplicación Spark, responsable de mantener información sobre la topología del clúster, distribuir el código y los datos a los ejecutores, y coordinar la ejecución de tareas en los ejecutores. El controlador es donde se inicia la aplicación Spark y donde se ejecuta el código del usuario.
   - Spark Executor: Son los procesos que realizan la ejecución real de las tareas de una aplicación Spark. Cada ejecutor se ejecuta en un nodo del clúster y es responsable de ejecutar un conjunto de tareas asignadas por el controlador. Los ejecutores son responsables de mantener los datos en memoria y realizar el procesamiento real de los datos de la aplicación.

**RESPUESTA ?**
   El Catalyst Optimizer se encarga de optimizar y ejecutar consultas SQL en Spark SQL. Su función es realizar optimizaciones en tiempo de ejecución para mejorar el rendimiento de las consultas, excluyendo columnas no utilizadas, la ordenación de operaciones y la generación de planes de ejecución más eficientes.

Todas estas variables corresponden a un valor entero discreto; sin embargo, todas utilizan valores numéricos del tipo double (float) con decimales, lo que en estricto rigor, si bien no entorpecen

````python
# RESPUESTA N° 5

from pyspark.sql.types import StringType, IntegerType, FloatType

def cast_cols(df, cols_types):
    for col_name, new_type in cols_types.items():
        df = df.withColumn(col_name, df[col_name].cast(new_type))
    return df

cols_to_cast = {"Quantity": IntegerType(), "Customer Age": IntegerType(),"Is Fraudulent": IntegerType(),"Account Age Days": IntegerType(),"Transaction Hour":IntegerType()}
df = cast_cols(df, cols_to_cast)
df.printSchema()
````

````python
# RESPUESTA N° 5

from pyspark.sql.functions import when, rand, col, sum

exprs = [sum(when(col(c).isNull(), 1).otherwise(0)).alias(c) for c in df.columns]

df.agg(*exprs).show()
````

````python
# RESPUESTA N° 3

df = df.withColumn('Type_of_purchase', when(df['Quantity'] < 5, 'minorista').otherwise('mayorista'))

df = df.withColumn('Is Alien', when(df['Is Alien'] == 1, 'Alien').otherwise('Human'))

df.show(5)
````

````python
# RESPUESTA N° 4

df.select(['Is Alien','Type_of_purchase','Transaction ID']).filter((df['Is Alien'] == 'Alien') &
          (df['Type_of_purchase'] == 'mayorista')).show()
````

````python
# RESPUESTA N° 5

df.groupBy("Is Alien").agg({"Quantity":"sum"}).show()
````

````python
# RESPUESTA N° 6

pivot_df = df.groupBy("Payment Method").pivot("Is Alien").agg({"Transaction Amount": "sum"})
pivot_df.show()
````

Además, están interesados en aprovechar las funciones de ventana en SQL para análisis avanzados. Las funciones de ventana permiten realizar cálculos sobre un conjunto de filas que están relacionadas con la fila actual. Por ejemplo, UNBOUNDED PRECEDING se usa para indicar que el rango de la función de ventana comienza desde la primera fila de la partición o del conjunto de resultados, lo cual es útil para calcular sumas acumulativas hasta la fila actual. Las variaciones comunes de este uso incluyen:

- `UNBOUNDED PRECEDING` to `CURRENT ROW`: Calcula desde el inicio de la partición hasta la fila actual.
- `UNBOUNDED PRECEDING` to `UNBOUNDED FOLLOWING`: Cubre todas las filas dentro de la partición.
- `VALUE PRECEDING` to `VALUE FOLLOWING`: Establece un rango específico basado en valores antes y después de la fila actual.

Ejemplo de uso en SQL:

```sql
STAT(COL1_NAME) OVER (PARTITION BY COL2_NAME ORDER BY COL3_NAME ROWS BETWEEN X PRECEDING AND CURRENT ROW)
```
